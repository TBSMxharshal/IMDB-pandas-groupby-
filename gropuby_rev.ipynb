```python
import pandas as pd
import numpy as np
```

```python
movies  = pd.read_csv("imdb-top-1000.csv")
movies.head()
```

```python
genres=movies.groupby("Genre")
```

```python
genres.sum()
```

```python
genres.min()
```

```python
genres.min()['Gross'].sort_values(ascending=False).head()
```

```python
#find the genre with highest avg imdb rating
genres["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# director with most popularity
movies.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(1)
```

```python
#find the highest rated movies of each genre
movies.groupby("Genre")[["IMDB_Rating", "Series_Title"]].max()
```

```python
# find no of movies done by each actor
movies["Star1"].value_counts()
```

```python
# by groupby mtd
movies.groupby("Star1")["Series_Title"].count().sort_values(ascending=False)
```

```python
#find total no of groups formed
len(movies.groupby("Genre"))
```

```python
# first movies of each genre
movies.groupby("Genre")["Series_Title"].first()
```

```python
# no of unique movies in each genre
genres.nunique()
```

```python
# doing different mathamatical expression on the columns
genres.agg({
    "IMDB_Rating":  "mean",
    "Gross": "sum",
    "No_of_Votes": "sum",
    "Runtime" : "mean"
}

)
```

```python
# min max mean for each coulumn
genres.agg(["min", "max"])
```

```python
#LOOPING ON GROUPS
for group,data in genres:
  print(data[data["IMDB_Rating"]==data["IMDB_Rating"].max()])


```

```python
# min value from each genre by using apply function
genres.apply(min)
```

```python
# find no of movies starting with A for each genre
def count_movies_starting_with_A(group):
    return (group["Series_Title"].str.startswith("A")).sum()


result = genres.apply(count_movies_starting_with_A)
print(result)
```

```python
# find ranking of each movie in the group according to the IMDB score
def ranking(group):
  group["Group rank"]=group["IMDB_Rating"].rank(ascending=False)
  return group

genres.apply(ranking)


```

```python
# find normalize ratring of each movie
def normal(group):
  group["norm_rating"] = ((group['IMDB_Rating'] - data['IMDB_Rating'].min()) / (data['IMDB_Rating'].max() - data['IMDB_Rating'].min()))
  return group
```

```python
genres.apply(normal)
```

```python
# find the most earning actor director combo
combo = movies.groupby(["Director","Star1"])
combo.sum().sort_values(by="Gross",ascending=False).head(1)
```

```python
# find the best (in-terms of meta score(avg) )actor-> genre combo
meta = movies.groupby(["Star1","Genre"])
meta["Metascore"].mean().sort_values(ascending=False).head(1)
```
